```python
# import modules
from PIL import Image
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# make sure you draw using white brush on black background
# in my experience, what seems to work best is using the thin white brush on MS Paint on a 28x28 black background

filepath = 'test.png'
im = Image.open(filepath, 'r')
px = im.load()
```

```python
# make sure it's the right image
im.show()
```

```python
# make sure you're working with square images, with size 28n x 28n so it can scale down easily to 28x28
width, height = im.size
print(width, height)
assert width == height
assert width % 28 == 0
```

```python
# convert image into 28x28 array of pixel values (0 is black, 255 is white)
scale = int(width / 28)
arr = np.zeros((28,28))
for i in range(width):
    for j in range(height):
        x = i // scale
        y = j // scale
        arr[y][x] += px[i,j][0]
arr = np.divide(arr, scale**2)
arr
```

```python
# plot the array as a sanity check
# make sure it resembles your digit and its 28x28
f, ax = plt.subplots(figsize=(9, 9))
sns.heatmap(arr, annot=True, ax=ax, square=True)
plt.show()
```

```python
# save this array into a csv file
np.savetxt('lata.csv', arr, delimiter=',')
```
